#### kopo_customerData 데이터에서 gender==male 이면서 email>0인 데이터를 추출하세요.

```python
kopo_customerData = pd.read_csv("./dataset/kopo_customerData.csv")
kopo_customerData.head()
```

```python
kopo_customerData.dtypes
```

```python
kopo_customerData = kopo_customerData [ (kopo_customerData.GENDER=="Male") & (kopo_customerData.EMAIL>0)]
kopo_customerData.head()
```

#### kopo_product_volume 데이터에서 productgroup==ST0002 이면서 VOLUME>300000 데이터를 추출하시오

```python
kopo_product_volume = pd.read_csv("./dataset/kopo_product_volume.csv")
kopo_product_volume.head()
```

```python
kopo_product_volume.dtypes
```

```python
kopo_product_volume.head()
```

```python
kopo_product_volume = kopo_product_volume [ (kopo_product_volume.PRODUCTGROUP=="ST0002") & 
                                           (kopo_product_volume.VOL>300000)]
kopo_product_volume
```

```python
def customFunction(inputColumn):
    #inputColumn=201601
    yearValue=inputColumn/100
    yearValue=int(yearValue)
    return yearValue
```

```python
selloutdata1["YEAR"]=selloutdata1["YEARWEEK"].apply(customFunction)
```

```python
selloutdata1=pd.read_csv("./dataset/mergedata1_0413.csv")
```

```python
selloutdata1.head()
```

```python
groupkey2=["REGIONID","PRODUCT"]
```

```python
def groupRolling(onegroup):
    indexgroupdata=onegroup.reset_index(drop=True)
    indexgroupdata["MA_MA"]=indexgroupdata["NEW_QTY"].rolling(
    window=5,center=True,min_periods=1).mean()
    
    return indexgroupdata
```

```python
finaldata=selloutdata1.groupby(groupkey2).apply(groupRolling)
```

```python
finaldata.head()
```

```python
finaldata=finaldata.reset_index(drop=True)
```
